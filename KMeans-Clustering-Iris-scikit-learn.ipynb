```python
from sklearn import datasets
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np


iris = datasets.load_iris()

features = iris.feature_names
data = iris.data
targets = iris.target

# initialize the model
model = KMeans(n_clusters=3)
# fit the model to data
model.fit(data)
# classify the data
labels = model.predict(data)

# split data into separate features
sepal_length = data[:,0]
sepal_width = data[:,1]
petal_length = data[:,2]
petal_width = data[:,3]

# get cluster attributes
sepal_clusters = model.cluster_centers_[:,:2]
petal_clusters = model.cluster_centers_[:,2:]

sepal_x = sepal_clusters[:,0]
sepal_y = sepal_clusters[:,1]

petal_x = petal_clusters[:,0]
petal_y = petal_clusters[:,1]


# plot iris sepal data and learned labels with clusters
plt.scatter(sepal_length, sepal_width, c=labels, alpha=0.25)
plt.scatter(sepal_x, sepal_y, c='k')
plt.xlabel("Iris Sepal Width")
plt.ylabel("Iris Sepal Height")
plt.title("Iris Sepal Dimensions")
plt.show()

# plot iris petal data and learned labels with clusters
plt.scatter(petal_length, petal_width, c=labels, alpha=0.25)
plt.scatter(petal_x, petal_y, c='k')
plt.xlabel("Iris Petal Width")
plt.ylabel("Iris Petal Height")
plt.title("Iris Petal Dimensions")
plt.show()


# score the model
score = 0
for i in range(len(labels)):
    labels.sort()
    targets.sort()
    if labels[i] == targets[i]:
        score += 1 / len(labels)
        
print("Score: ", score)
```
